# Run Auction Experiment

Now we'll run an experiment using the newly created auction task

```python
from benchmarl.algorithms import MappoConfig
from benchmarl.environments import AuctionTask
from benchmarl.experiment import Experiment, ExperimentConfig
from benchmarl.models.mlp import MlpConfig

# Load configurations
experiment_config = ExperimentConfig.get_from_yaml()
task = AuctionTask.FIRST_PRICE.get_from_yaml()
algorithm_config = MappoConfig.get_from_yaml()
model_config = MlpConfig.get_from_yaml()
critic_model_config = MlpConfig.get_from_yaml()

# Modify experiment configuration for a quick test
experiment_config.max_n_iters = 5
experiment_config.on_policy_collected_frames_per_batch = 1000
experiment_config.loggers = []

# Create and run the experiment
experiment = Experiment(
    task=task,
    algorithm_config=algorithm_config,
    model_config=model_config,
    critic_model_config=critic_model_config,
    seed=0,
    config=experiment_config,
)
experiment.run()
```

You can modify the experiment parameters or try different algorithms and tasks as needed.
